## ðŸ“ ESERCIZIO: Gestione voti scolastici

Creiamo un sistema per gestire i voti di uno studente usando **dizionari** e **liste** insieme.

### Struttura dati
```python
{
    "Matematica": [8, 9, 7],
    "Italiano": [7, 8, 8, 9],
    "Inglese": [9, 9]
}
```

Ogni **chiave** Ã¨ una materia, ogni **valore** Ã¨ una lista di voti.

### Parte 1: Creare e visualizzare i voti

````python
# Dizionario con voti per materia
voti = {
    "Matematica": [8, 9, 7, 8],
    "Italiano": [7, 8, 8, 9],
    "Inglese": [9, 9, 10],
    "Storia": [6, 7, 7],
    "Scienze": [8, 9, 9, 10]
}

# Visualizzare tutti i voti
print("ðŸ“š REGISTRO VOTI")
print("="*40)
for materia, lista_voti in voti.items():
    print(f"{materia:12} â†’ {lista_voti}")
````

### Parte 2: Calcolare la media per materia

````python
# Calcolare la media per ogni materia
print("ðŸ“Š MEDIE PER MATERIA")
print("="*40)

for materia, lista_voti in voti.items():            # Enumera il dizionario
    media = sum(lista_voti) / len(lista_voti)
    print(f"{materia:12} â†’ Media: {media:.2f}")
````

### Parte 3: Un'altra soluzione piÃ¹ "pythonesca"

````python
# Creazione di un dizionario
medie = {materia: sum(lista_voti) / len(lista_voti) for materia, lista_voti in voti.items()}

print("ðŸ“Š MEDIE CALCOLATE")
for materia, media in medie.items():
    print(f"{materia:12} â†’ {media:.2f}")
````

### Parte 4: Calcolare la media complessiva

````python
# Approccio classico
num_voti = 0
totale_voti = 0
for lista_voti in voti.values():
  num_voti = num_voti + len(lista_voti)
  totale_voti = totale_voti + sum(lista_voti)

media_complessiva = totale_voti / num_voti
print(f"ðŸŽ¯ MEDIA COMPLESSIVA (da voti):  {media_complessiva:.2f}")

````

Una soluzione piÃ¹ avanzata che usa list comprehensions, ovvero liste create con una iterazione su elementi di altre liste

````python
# Approccio 2: da tutti i voti (piÃ¹ preciso)
tutti_i_voti = [voto for lista_voti in voti.values() for voto in lista_voti]
media_complessiva_v2 = sum(tutti_i_voti) / len(tutti_i_voti)

print(f"ðŸŽ¯ MEDIA COMPLESSIVA (da voti):  {media_complessiva_v2:.2f}")
print(f"\nðŸ“Œ Totale voti registrati: {len(tutti_i_voti)}")
````

### ðŸ§  Concetti pythonesco vs teorico

#### âŒ Approccio teorico-scolastico:
```python
# Troppo verboso, variabili inutili
somma = 0
contatore = 0
for voto in voti_matematica:
    somma = somma + voto
    contatore = contatore + 1
media = somma / contatore
```

#### âœ… Approccio pythonesco:
```python
# Conciso, leggibile, idiomatico
media = sum(voti_matematica) / len(voti_matematica)
```

#### ðŸš€ Ancora piÃ¹ pythonesco:
```python
# Con list/dict comprehension
medie = {mat: sum(v)/len(v) for mat, v in voti.items()}
```

**Principio**: usa le funzioni built-in di Python (`sum`, `len`, `max`, `min`) e le comprehension per codice piÃ¹ pulito!

## ðŸŽ¯ Filtraggio: Come con dizionari e liste

````python
# FILTRO 1: Voti di un singolo studente (come dizionario!)
studente = "Rossi"
voti_rossi = df_voti[df_voti['Cognome'] == studente]

print(f"ðŸ“š VOTI DI {studente}")
print("="*60)
print(voti_rossi[['Materia', 'Voto', 'Data']])

print(f"\nðŸŽ¯ Media {studente}: {voti_rossi['Voto'].mean():.2f}")
````

````python
# FILTRO 2: Voti sufficienti (come filtravamo le liste!)
# Prima: pari = [n for n in numeri if n % 2 == 0]
# Ora: voti sufficienti

sufficienti = df_voti[df_voti['Voto'] >= 6]
insufficienti = df_voti[df_voti['Voto'] < 6]

print(f"âœ… Voti sufficienti: {len(sufficienti)} ({len(sufficienti)/len(df_voti)*100:.1f}%)")
print(f"âŒ Voti insufficienti: {len(insufficienti)} ({len(insufficienti)/len(df_voti)*100:.1f}%)")
````

````python
# FILTRO 3: Combinare condizioni (AND, OR)
# Voti alti in Matematica
matematica_alta = df_voti[(df_voti['Materia'] == 'Matematica') & (df_voti['Voto'] >= 8)]

print("ðŸŒŸ ECCELLENZE IN MATEMATICA (voti >= 8)")
print("="*60)
print(matematica_alta[['Nome', 'Cognome', 'Voto', 'Data']].head(10))
````

## ðŸ“Š Statistiche Avanzate: Oltre i dizionari

````python
# TABELLA PIVOT: Medie per Studente e Materia
# Questo era impossibile con semplici dizionari!

pivot_medie = df_voti.pivot_table(
    values='Voto',
    index='Cognome',
    columns='Materia',
    aggfunc='mean'
).round(2)

pivot_count = df_voti.pivot_table(
    values='Voto',
    index='Cognome',
    columns='Materia',
    aggfunc='count'
).round(2)

print("ðŸ“Š TABELLA MEDIE: STUDENTE Ã— MATERIA")
print("="*80)
print(pivot_medie)
print(pivot_count)
````

````python
# STATISTICHE COMPLETE (come facevamo con i dizionari, ma automatico!)
print("ðŸ“ˆ STATISTICHE DETTAGLIATE PER MATERIA")
print("="*80)

stats = df_voti.groupby('Materia')['Voto'].agg([
    ('Media', 'mean'),
    ('Mediana', 'median'),
    ('Min', 'min'),
    ('Max', 'max'),
    ('Num_Voti', 'count')
]).round(2)

print(stats)
````

## ðŸ’ª ESERCIZIO PRATICO: Collega dizionari e Pandas

Prova a rispondere a queste domande usando Pandas (come facevamo con i dizionari!):

1. Qual Ã¨ la materia piÃ¹ difficile (media piÃ¹ bassa)?
2. Qual Ã¨ la materia piÃ¹ facile (media piÃ¹ alta)?
3. Quale studente ha la media piÃ¹ alta?
4. Quanti voti ha preso ogni studente?
5. In quale classe si trova lo studente con la media piÃ¹ alta?

````python
# SOLUZIONI

# 1. Materia piÃ¹ difficile
materia_difficile = df_voti.groupby('Materia')['Voto'].mean().idxmin()
media_difficile = df_voti.groupby('Materia')['Voto'].mean().min()
print(f"ðŸ“‰ Materia piÃ¹ difficile: {materia_difficile} (media: {media_difficile:.2f})")

# 2. Materia piÃ¹ facile
materia_facile = df_voti.groupby('Materia')['Voto'].mean().idxmax()
media_facile = df_voti.groupby('Materia')['Voto'].mean().max()
print(f"ðŸ“ˆ Materia piÃ¹ facile: {materia_facile} (media: {media_facile:.2f})")

# 3. Studente con media piÃ¹ alta
medie_studenti = df_voti.groupby('Cognome')['Voto'].mean()
studente_top = medie_studenti.idxmax()
media_top = medie_studenti.max()
print(f"ðŸ† Studente top: {studente_top} (media: {media_top:.2f})")

# 4. Numero voti per studente
voti_count = df_voti.groupby('Cognome').size()
print(f"\nðŸ“ Voti per studente:")
print(voti_count.head())

# 5. Classe dello studente top
classe_top = df_voti[df_voti['Cognome'] == studente_top]['Classe'].iloc[0]
print(f"\nðŸŽ“ {studente_top} Ã¨ nella classe {classe_top}")
````

## Avanzato

I for si possono usare per definire nuove liste ("list comprehension").
Se vi ricordate in matematica la definizione di insiemi di elementi che verificano una proprietÃ  la sintessi sarÃ  piÃ¹ facile.

$$
\text{Pari} = \{ n \in \mathbb{N} \mid n \equiv 0 \pmod{2} \}
$$

$$
\mathbb{P} = \{ p \in \mathbb{N} \mid \forall k \in \mathbb{N}, 2 \leq k < p \Rightarrow p \not\equiv 0 \pmod{k} \}
$$



````python
quadrati = [i**2 for i in range(1,6)]

# Posso anche aggiungere elementi nella lista solo se verificano una condizione.
pari= [k for k in range(1,11) if k % 2 == 0]

# esempio illustrativo, lo Zen del Python dice "simpler is better"
# Ã¨ meglio scrivere [k for k in range(2,11,2)]

# Esempio avanzato, numeri primi con list comprehension
primi = [p for p in range(1,11) if all(p % k != 0 for k in range(2,p))]

print(f"Quadrati: {quadrati}")
print(f"Pari: {pari}")
print(f"Primi: {primi}")
````

## ðŸ“Š ESERCIZIO: Visualizzazione dati con Pandas

Pandas integra Matplotlib per creare grafici direttamente dai DataFrame.
Vediamo come rappresentare i voti in modo visuale.


### Parte 1: Preparazione dati


````python
import pandas as pd
import matplotlib.pyplot as plt

# Dati voti per materia
voti = {
    "Matematica": [8, 9, 7, 8, 9],
    "Italiano": [7, 8, 8, 9, 7],
    "Inglese": [9, 9, 10, 8, 9],
    "Storia": [6, 7, 7, 8, 6],
    "Scienze": [8, 9, 9, 10, 8]
}

# Creo DataFrame
df_voti = pd.DataFrame(voti)
print("DataFrame voti:")
print(df_voti)
print(f"\nDimensioni: {df_voti.shape}")
````

### Parte 3: Grafico a linee - Andamento temporale


````python
# Grafico a linee per vedere l'andamento dei voti nel tempo
plt.figure(figsize=(12, 6))
df_voti.plot(marker='o', linewidth=2)
plt.title('Andamento voti nel tempo', fontsize=16, fontweight='bold')
plt.xlabel('Verifica numero', fontsize=12)
plt.ylabel('Voto', fontsize=12)
plt.ylim(5, 11)
plt.legend(title='Materie', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
````

````python
# Il tuo codice qui!

# 1. Aggiungi Fisica
df_voti['Fisica'] = [7, 8, 9, 8, 7]


# 3. Confronto min, media, max
confronto = pd.DataFrame({
    'Minimo': df_voti.min(),
    'Media': df_voti.mean(),
    'Massimo': df_voti.max()
})

plt.figure(figsize=(12, 6))
confronto.plot(kind='bar', edgecolor='black')
plt.title('Confronto Min-Media-Max per materia', fontsize=14, fontweight='bold')
plt.xlabel('Materia')
plt.ylabel('Voto')
plt.xticks(rotation=45)
plt.legend(title='Statistica')
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
````
